# Functions

```python
# === Generate 3D Gaussian Random Field (GRF) ===
def generate_3d_grf(shape, correlation_length=0.8, physical_size=4):
    """Generates a 3D Gaussian random field with a Gaussian power spectrum."""
    
    # Create 3D white noise
    noise = np.random.normal(size=shape)

    # Define physical frequency grid
    Lx, Ly, Lz = physical_size, physical_size, physical_size  # Size of the domain
    kx = np.fft.fftfreq(shape[0], d=Lx / shape[0])[:, None, None]
    ky = np.fft.fftfreq(shape[1], d=Ly / shape[1])[None, :, None]
    kz = np.fft.fftfreq(shape[2], d=Lz / shape[2])[None, None, :]
    k = np.sqrt(kx**2 + ky**2 + kz**2)  # Radial frequency grid

    # Define Gaussian power spectrum
    k_c = 1.0 / correlation_length  # Characteristic wavenumber
    spectrum = np.exp(-0.5 * (k / k_c) ** 2)  # Gaussian decay
    spectrum[0, 0, 0] = 0  # Remove mean shift (zero-frequency mode)

    # Apply filter in Fourier space
    filtered_noise = fftn(noise) * spectrum
    grf = np.real(ifftn(filtered_noise))  # Transform back to real space

    # Normalize to zero mean and unit variance
    grf -= np.mean(grf)
    grf /= np.std(grf)

    return grf
```

## Convex spheroid

### Mayavi

```python
bool_save = True
# === Adjustable parameters ===
R1 = 0.9  # Radius of dumbbell neck
L1 = 1  # Half-length of dumbbell
k1 = 0.05  # Asymmetry parameter
R2 = 0.05  # Radius of sphere
offset2 = (0,1.75,-1.75)  # Offset of sphere from origin
color2 = (1, 0.3, 0.3)  # Color of sphere
R3 = 0.05  # Radius of sphere
offset3 = (0.5,-0.5,1)  # Offset of sphere from origin
color3 = (0.3, 0.3, 1)  # Color of sphere

# Define grid for angles
theta = np.linspace(0, np.pi, 500)  # Polar angle
phi = np.linspace(0, 2 * np.pi, 500)  # Azimuthal angle
theta, phi = np.meshgrid(theta, phi)

# Define the 3D Cartesian grid that covers the whole region
grid_size = (300, 300, 300)  # Resolution of the 3D GRF
x_lin = np.linspace(-L1, L1, grid_size[0])  # X range
y_lin = np.linspace(-L1, L1, grid_size[1])  # Y range
z_lin = np.linspace(-L1, L1, grid_size[2])  # Z range

# === Parametric equations for dumbbell (1) and sphere (2) ===
rho1 = (L1 * np.sin(theta) - (L1 - R1) * np.sin(theta)**2) + k1 * np.heaviside(theta-np.pi/2,0) * np.sin(2*theta)**2
X1 = rho1 * np.cos(phi)
Y1 = rho1 * np.sin(phi)
Z1 = L1 * np.cos(theta)
X2 = R2 * np.sin(theta) * np.cos(phi) + offset2[0]
Y2 = R2 * np.sin(theta) * np.sin(phi) + offset2[1]
Z2 = R2 * np.cos(theta) + offset2[2]
X3 = R3 * np.sin(theta) * np.cos(phi) + offset3[0]
Y3 = R3 * np.sin(theta) * np.sin(phi) + offset3[1]
Z3 = R3 * np.cos(theta) + offset3[2]

grf_3d = generate_3d_grf(grid_size, correlation_length=0.5)
interp = RegularGridInterpolator((x_lin, y_lin, z_lin), grf_3d, bounds_error=False, fill_value=0)
grf1 = interp(np.stack([X1, Y1, Z1], axis=-1))  # Interpolate the GRF onto surface

# === Render the surfaces ===
fig = mlab.figure(size=(1024, 1024), bgcolor=(1, 1, 1))  # White background
mesh1 = mlab.mesh(X1, Y1, Z1, scalars=grf1, opacity=1)
mesh2 = mlab.mesh(X2, Y2, Z2, color=color2, opacity=0.9)
mesh3 = mlab.mesh(X3, Y3, Z3, color=color3, opacity=0.9)

# === Load Colormap ===
cmap = cm.gray
lut_new = (cmap(np.linspace(0, 1, 256))[:, :] * 255).astype(np.uint8)
mesh1.module_manager.scalar_lut_manager.lut.table = lut_new  # Apply LUT to surface

# === Apply Raymond-style shading ===
for mesh in [mesh1, mesh2, mesh3]:
    actor = mesh.actor
    actor.property.interpolation = 'phong'  # Smooth shading
    actor.property.specular = 0.2  # Specular highlights
    actor.property.specular_power = 20  # Sharpness of highlights
    actor.property.ambient = 0.1 if mesh == mesh1 else 1  # Adjust ambient light
    actor.property.diffuse = 1  # Full diffuse lighting

# === Add black lines parallel to the z-axis ===
line_positions = [(0, 0), (offset2[0], offset2[1]), (offset3[0], offset3[1])]
z_min, z_max = -2.2 * L1, 2.2 * L1  # Adjust range to cover both meshes

for x_pos, y_pos in line_positions:
    z_vals = np.linspace(z_min, z_max, 100)
    x_vals = np.full_like(z_vals, x_pos)
    y_vals = np.full_like(z_vals, y_pos)
    mlab.plot3d(x_vals, y_vals, z_vals, color=(0, 0, 0), tube_radius=0.01)  # Black lines

# Force update of the figure
mlab.draw()
mlab.view(azimuth=0, elevation=90, roll=-45, distance=10)
fig.scene._lift()  # This needs to be called for light_manager to work

# Scene lights (4 lights are used)
engine = mlab.get_engine()
scene = engine.current_scene

def set_lights(intensities):
    cam_light_azimuth = [-90, 90, -45, 0]
    cam_light_elevation = [45, 0, 10, 0]
    cam_light_color = [(1, 1, 1), color2, color3, (1, 1, 1)]  # White, green, white, white
    
    for i in range(4):
        cam_light = scene.scene.light_manager.lights[i]
        cam_light.activate = True
        cam_light.azimuth = cam_light_azimuth[i]
        cam_light.elevation = cam_light_elevation[i]
        cam_light.intensity = intensities[i]
        cam_light.color = cam_light_color[i]

# mlab.orientation_axes()

# === Step 1: Set initial light intensities and save as explosion_1.png ===
set_lights([1, 0, 0, 0])
if bool_save:
    mlab.savefig("explosion_convex_1.png", figure=fig, magnification=2)

# === Step 2: Change lights to new intensities and save as explosion_2.png, etc ===
# === Change photosphere Colormap to gray
cmap = cm.gray
lut_new = (cmap(np.linspace(0, 1, 256))[:, :] * 255).astype(np.uint8)
mesh1.module_manager.scalar_lut_manager.lut.table = lut_new  # Apply LUT to surface
if bool_save:
    set_lights([0, 1, 0, 0])
    mlab.savefig("explosion_convex_2.png", figure=fig, magnification=2)
    set_lights([0, 0, 1, 0])
    mlab.savefig("explosion_convex_3.png", figure=fig, magnification=2)

# Show the result
mlab.show()
```
