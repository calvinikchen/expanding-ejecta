## Photosphere

```python
def I_lambda_ph(theta,theta_ph=1):
    cos_psi = np.sqrt(1 - theta**2 / theta_ph**2, where=theta < theta_ph, out=np.ones_like(theta))
    a_1 = 0.93
    a_2 = -0.23
    a_0 = 1 - a_1 - a_2
    return np.heaviside(1-theta/theta_ph,0) * (a_0 + a_1 * cos_psi + a_2 * cos_psi**2)
```

## Ejecta

```python
# === Generate 3D Gaussian Random Field (GRF) ===
def generate_3d_grf(shape, correlation_length=0.8, physical_size=(2,2,6)):
    """Generates a 3D Gaussian random field with a Gaussian power spectrum."""
    
    # Create 3D white noise
    noise = np.random.normal(size=shape)

    # Define physical frequency grid
    Lx, Ly, Lz = physical_size[0], physical_size[1], physical_size[2]  # Size of the domain
    kx = np.fft.fftfreq(shape[0], d=Lx / shape[0])[:, None, None]
    ky = np.fft.fftfreq(shape[1], d=Ly / shape[1])[None, :, None]
    kz = np.fft.fftfreq(shape[2], d=Lz / shape[2])[None, None, :]
    k = np.sqrt(kx**2 + ky**2 + kz**2)  # Radial frequency grid

    # Define Gaussian power spectrum
    k_c = 1.0 / correlation_length  # Characteristic wavenumber
    spectrum = np.exp(-0.5 * (k / k_c) ** 2)  # Gaussian decay
    spectrum[0, 0, 0] = 0  # Remove mean shift (zero-frequency mode)

    # Apply filter in Fourier space
    filtered_noise = fftn(noise) * spectrum
    grf = np.real(ifftn(filtered_noise))  # Transform back to real space

    # Normalize to zero mean and unit variance
    grf -= np.mean(grf)
    grf /= np.std(grf)

    return grf
```

```python
def grf_interpolate(grf, physical_size, t, X, Y, Z):
    """
    Interpolates the 3D Gaussian Random Field (GRF) at rescaled coordinates X/t, Y/t, Z_0 + Z-Z_0/t, where Z_0 = 1/3 * physical_size[2]
    
    Parameters:
        grf (ndarray): 3D array representing the Gaussian Random Field.
        physical_size (tuple): The physical dimensions (Lx, Ly, Lz) of the GRF domain.
        t (float): Scaling factor.
        X, Y, Z (ndarray): Coordinate grids to evaluate the interpolated GRF.
    
    Returns:
        ndarray: Interpolated GRF values at the transformed coordinates.
    """
    # Define the grid points based on the GRF shape and physical size
    nx, ny, nz = grf.shape
    x = np.linspace(-physical_size[0] / 2, physical_size[0] / 2, nx)
    y = np.linspace(-physical_size[1] / 2, physical_size[1] / 2, ny)
    z = np.linspace(-physical_size[2] / 2, physical_size[2] / 2, nz)
    Z_0 = 1/3 * physical_size[2]
    
    # Create the interpolator
    interpolator = RegularGridInterpolator((x, y, z), grf, bounds_error=False, fill_value=0)
    
    # Rescale the input coordinates
    X_scaled = X / t
    Y_scaled = Y / t
    Z_scaled = Z_0 + (Z-Z_0) / t

    # Stack coordinates for interpolation
    points = np.vstack([X_scaled.ravel(), Y_scaled.ravel(), Z_scaled.ravel()]).T

    # Interpolate the values and reshape back to original shape
    grf_interpolated = interpolator(points).reshape(X.shape)
    
    return grf_interpolated
```

```python
def generate_tau_ejecta(r, grf, tau_0=2, delta_tau=0.2, spectral_index=4, theta_ph=1, t=1):
    """Generates the ejecta tau profile."""
    smooth_profile = np.heaviside(r - theta_ph, 0) * tau_0 * ((r/t)/theta_ph+1e-10)**(-spectral_index)
    
    # Define the ejecta intensity profile
    tau_ejecta = smooth_profile * (1 + delta_tau * grf)
    
    return tau_ejecta
```

```python
def dilution_factor(r, theta_ph=1):
    return (1/2) * (1 - np.sqrt(1 - theta_ph**2 / (r**2+1e-15), where= r > theta_ph, out=np.zeros_like(r)))
```

```python
def I_lambda_absorption(X, Y, Z, grf, tau_0=2, delta_tau=0.2, spectral_index=4, theta_ph=1, t=1):
    """Generates the absorption intensity profile."""
    theta = np.sqrt(X**2 + Y**2)
    r = np.sqrt(X**2 + Y**2 + Z**2)
    tau_ejecta = generate_tau_ejecta(r, grf, tau_0, delta_tau, spectral_index, theta_ph, t)
    frontside = np.heaviside(-Z,0) # only consider the front side of the ejecta
    return I_lambda_ph(theta, theta_ph) * (1 - (1-np.exp(-tau_ejecta)) * frontside)
```

```python
def I_lambda_emission(X, Y, Z, grf, tau_0=2, delta_tau=0.5, spectral_index=4, theta_ph=1, t=1, occultation=True):
    r = np.sqrt(X**2 + Y**2 + Z**2)
    theta = np.sqrt(X**2 + Y**2)
    tau_ejecta = generate_tau_ejecta(r, grf, tau_0, delta_tau, spectral_index, theta_ph, t)
    dilution = dilution_factor(r, theta_ph)
    
    if occultation:
        frontside = np.heaviside(-Z,1) # front side of the ejecta
        backside = np.heaviside(Z,0) # back side of the ejecta
        return dilution * (1 - np.exp(-tau_ejecta)) * (frontside + backside * np.heaviside(theta - theta_ph,1))
    else:
        return dilution * (1 - np.exp(-tau_ejecta)) 
```

```python
def I_lambda_total(X, Y, Z, grf, tau_0=2, delta_tau=0.5, spectral_index=4, theta_ph=1, t=1):
    return I_lambda_emission(X, Y, Z, grf, tau_0, delta_tau, spectral_index, theta_ph, t) + I_lambda_absorption(X, Y, Z, grf, tau_0, delta_tau, spectral_index, theta_ph, t)
```

```python
# Function to add colorbars
def add_colorbar(ax, im, label="", size="5%"):
    divider = make_axes_locatable(ax)
    cax = divider.append_axes("right", size=size, pad=0.05)
    cbar = plt.colorbar(im, cax=cax)
    #cbar.set_label(label, fontsize=12)
```

```python
'''
    == A few notes about color ==
    Color   Wavelength(nm) Frequency(THz)
    Red     620-750        484-400
    Orange  590-620        508-484
    Yellow  570-590        526-508
    Green   495-570        606-526
    Blue    450-495        668-606
    Violet  380-450        789-668
    f is frequency (cycles per second)
    l (lambda) is wavelength (meters per cycle)
    e is energy (Joules)
    h (Plank's constant) = 6.6260695729 x 10^-34 Joule*seconds
                         = 6.6260695729 x 10^-34 m^2*kg/seconds
    c = 299792458 meters per second
    f = c/l
    l = c/f
    e = h*f
    e = c*h/l
    List of peak frequency responses for each type of 
    photoreceptor cell in the human eye:
        S cone: 437 nm
        M cone: 533 nm
        L cone: 564 nm
        rod:    550 nm in bright daylight, 498 nm when dark adapted. 
                Rods adapt to low light conditions by becoming more sensitive.
                Peak frequency response shifts to 498 nm.
'''

def wavelength_to_rgb(wavelength, gamma=0.8):

    '''This converts a given wavelength of light to an 
    approximate RGB color value. The wavelength must be given
    in nanometers in the range from 380 nm through 750 nm
    (789 THz through 400 THz).
    Based on code by Dan Bruton
    http://www.physics.sfasu.edu/astro/color/spectra.html
    '''

    wavelength = float(wavelength)
    if wavelength >= 380 and wavelength <= 440:
        attenuation = 0.3 + 0.7 * (wavelength - 380) / (440 - 380)
        R = ((-(wavelength - 440) / (440 - 380)) * attenuation) ** gamma
        G = 0.0
        B = (1.0 * attenuation) ** gamma
    elif wavelength >= 440 and wavelength <= 490:
        R = 0.0
        G = ((wavelength - 440) / (490 - 440)) ** gamma
        B = 1.0
    elif wavelength >= 490 and wavelength <= 510:
        R = 0.0
        G = 1.0
        B = (-(wavelength - 510) / (510 - 490)) ** gamma
    elif wavelength >= 510 and wavelength <= 580:
        R = ((wavelength - 510) / (580 - 510)) ** gamma
        G = 1.0
        B = 0.0
    elif wavelength >= 580 and wavelength <= 645:
        R = 1.0
        G = (-(wavelength - 645) / (645 - 580)) ** gamma
        B = 0.0
    elif wavelength >= 645 and wavelength <= 750:
        attenuation = 0.3 + 0.7 * (750 - wavelength) / (750 - 645)
        R = (1.0 * attenuation) ** gamma
        G = 0.0
        B = 0.0
    else:
        R = 0.0
        G = 0.0
        B = 0.0
    # R *= 255
    # G *= 255
    # B *= 255
    # return (int(R), int(G), int(B))
    return (R, G, B)
```

```python
def cmap_black_to_rgb(rgb_color):
    """Creates a custom colormap transitioning from black to a specified RGB color.

    Args:
        rgb_color: A tuple representing the RGB color (e.g., (1, 0, 0) for red).

    Returns:
        A matplotlib LinearSegmentedColormap object.
    """
    cdict = {'red':   [[0.0, 0.0, 0.0],
                       [1.0, rgb_color[0], rgb_color[0]]],
             'green': [[0.0, 0.0, 0.0],
                       [1.0, rgb_color[1], rgb_color[1]]],
             'blue':  [[0.0, 0.0, 0.0],
                       [1.0, rgb_color[2], rgb_color[2]]]}
    return mcolors.LinearSegmentedColormap('black_to_rgb', cdict)
```

```python
# Define time range
t_values = np.linspace(0.6, 1.8, num=61)  # Adjust number of frames as needed
bool_save = True  

# Set parameters
theta_ph = 0.4
physical_size = (2, 2, 6)
tau_0 = 2
delta_tau = 0.4
correlation_length = 1
spectral_index = 4
n_idx = 300
idx_z_sample = 5
freq_slice = 200
n_pad = 5 * n_idx
n_mid = int(2*n_pad + n_idx)//2
shape = (n_idx+1, n_idx+1, 3*n_idx+1)
x_lin = np.linspace(-physical_size[0]/2, physical_size[0]/2, shape[0])
y_lin = np.linspace(-physical_size[1]/2, physical_size[1]/2, shape[1])
z_lin = np.linspace(-(2/3)*physical_size[2], (1/3)*physical_size[2], shape[2])
X, Y, Z = np.meshgrid(x_lin, y_lin, z_lin, indexing='ij')
grf = generate_3d_grf(shape, correlation_length=correlation_length, physical_size=physical_size)

# Slices
idx_mid_proj = 2*n_idx +1
idx_mid_side = n_idx // 2 +1
idx_v_z_0 = idx_mid_proj - int(1 * n_idx)
idx_v_z_1 = idx_mid_proj - int(0.20 * n_idx)
idx_v_z_2 = idx_mid_proj + int(0.1 * n_idx)

# Setup figure
fig = plt.figure(figsize=(16.5, 6.5))
fig.subplots_adjust(right=0.9, wspace=0.2, hspace=0) # make space for colorbar
gs = gridspec.GridSpec(2, 4, width_ratios=[1, 1, 1, 3], height_ratios=[1, 1])

ax00 = fig.add_subplot(gs[0, 0])  
ax01 = fig.add_subplot(gs[0, 1])  
ax02 = fig.add_subplot(gs[0, 2])  
ax03 = fig.add_subplot(gs[0, 3])  
ax10 = fig.add_subplot(gs[1, 0])  
ax11 = fig.add_subplot(gs[1, 1])  
ax12 = fig.add_subplot(gs[1, 2])  
ax13 = fig.add_subplot(gs[1, 3])  

# Add annotations
ax03.text(-1.3,+0.17,r"absorption",color=(1,1,1,1),ha='center',va='center')
ax03.text(-1.3,0,r"+",color=(1,1,1,1),ha='center',va='center')
ax03.text(-1.3,-0.17,r"emission",color=(1,1,1,1),ha='center',va='center')
ax03.text(-1.3,-0.6,r"emission",color=(1,1,1,1),ha='center',va='center')
ax03.text(-1.3,+0.6,r"emission",color=(1,1,1,1),ha='center',va='center')
ax03.text(1,0,r"occulted",color=(1,1,1,0.6))
ax03.annotate('blueshifted',color=(0.5,0.5,1,1), xy=(-0.02, -0.9), xytext=(-1, -0.9), ha='right', va='center', arrowprops=dict(color=(0.5,0.5,1,1),arrowstyle="<-",lw=1))
ax03.annotate('redshifted',color=(1,0.5,0.5,1), xy=(0.02, -0.9), xytext=(1, -0.9), ha='left', va='center', arrowprops=dict(color=(1,0.5,0.5,1),arrowstyle="<-",lw=1))
eye_triangle = Line2D([-3.8,-3.95,-3.8], [0.15,0,-0.15], color='white', lw=1)
eye_ball = Arc([-3.95, 0], width=0.35, height=0.35, angle=0, theta1=-45, theta2=45, color='white', lw=1)
eye_pupil = Arc([-3.72, 0], width=0.25, height=0.45, angle=0, theta1=130, theta2=-130, color='white', lw=1)
pupil = Ellipse([-3.79, 0], width=0.03, height=0.12, color='white')

# List of axes and titles
list_ax = [ax00, ax01, ax02, ax03, ax10, ax11, ax12, ax13]
list_titles = [
    r'$I_\lambda(\lambda_1,\boldsymbol{\theta})$', 
    r'$I_\lambda(\lambda_2,\boldsymbol{\theta})$', 
    r'$I_\lambda(\lambda_3,\boldsymbol{\theta})$',
    r'$I_\lambda(\lambda_\mathrm{rest},\mathrm{side~view})$',
    r'$|\mathcal{V}(\lambda_1,\boldsymbol{u})|^2$', 
    r'$|\mathcal{V}(\lambda_2,\boldsymbol{u})|^2$', 
    r'$|\mathcal{V}(\lambda_3,\boldsymbol{u})|^2$', 
    r'$t = %.2g$' % t]
list_aspect = [1, 1, 1, 1, 1, 1, 1, 1]  # Aspect ratio for each subplot
list_x_labels = [r"$\theta_x$", r"$\theta_x$", r"$\theta_x$", r"$z = v_\parallel (t-t_0)$",
                 r"$u$", r"$u$", r"$u$", r"$\lambda = \lambda_\mathrm{rest} (1 + v_\parallel)$"]
list_y_labels = [r"$\theta_y$", "", "", "",
                 r"$v$", "", "", r"$f_\lambda$"]
x_ticks = np.linspace(-physical_size[0]/2, physical_size[0]/2, 5)
y_ticks = np.linspace(-physical_size[1]/2, physical_size[1]/2, 5)
f_ticks = [0]
list_x_ticks = [[], [], [], [], [], [], [], []]
list_y_ticks = [[], [], [], [], [], [], [], f_ticks]

for ax, title, aspect, x_label, y_label in zip(list_ax, list_titles, list_aspect, list_x_labels, list_y_labels):
    ax.set_title(title, fontsize=16, pad=10)  # Set consistent padding for alignment
    ax.set_aspect(aspect)  # Set aspect ratio
    ax.set_xlabel(x_label)
    ax.set_ylabel(y_label)
    ax.set_xticks(list_x_ticks[list_ax.index(ax)],minor=False)
    ax.set_yticks(list_y_ticks[list_ax.index(ax)],minor=False)

# Set limits and positions
ax03.set_xlim(-(2/3)*physical_size[2], (1/3)*physical_size[2])
ax03.set_ylim(-physical_size[1]/2, physical_size[1]/2)
ax13.set_xlim(-(2/3)*physical_size[2], (1/3)*physical_size[2])
ax13.set_ylim(0,2)
pos03 = ax03.get_position()
pos13 = ax13.get_position()
ax03.set_position([pos03.x0, pos03.y0+0.012, pos03.width, pos03.height-0.024]) # Force ax03 to align with ax02 while keeping its width
ax13.set_position([pos13.x0, pos13.y0+0.012, pos13.width, pos13.height-0.024]) # Force ax13 to align with ax02 while keeping its width

# Initialize placeholders for images
im00 = ax00.pcolormesh(X[:,:,idx_v_z_0], Y[:,:,idx_v_z_0], np.zeros((shape[0],shape[1])), cmap=cmap_0, norm=mcolors.LogNorm(vmin=3e-2, vmax=1), rasterized=True)
im01 = ax01.pcolormesh(X[:,:,idx_v_z_1], Y[:,:,idx_v_z_1], np.zeros((shape[0],shape[1])), cmap=cmap_1, norm=mcolors.LogNorm(vmin=3e-2, vmax=1), rasterized=True)
im02 = ax02.pcolormesh(X[:,:,idx_v_z_2], Y[:,:,idx_v_z_2], np.zeros((shape[0],shape[1])), cmap=cmap_2, norm=mcolors.LogNorm(vmin=3e-2, vmax=1), rasterized=True)
im03 = ax03.pcolormesh(Z[idx_mid_side,:,:], Y[idx_mid_side,:,:], np.zeros((shape[1], shape[2])), cmap=cm.inferno, vmin=0, vmax=1, rasterized=True)
im10 = ax10.imshow(np.zeros((freq_slice,freq_slice)), cmap=cmap_0,norm=mcolors.LogNorm(vmin=1e-4, vmax=1),rasterized=True)
im11 = ax11.imshow(np.zeros((freq_slice,freq_slice)), cmap=cmap_1,norm=mcolors.LogNorm(vmin=1e-4, vmax=1),rasterized=True)
im12 = ax12.imshow(np.zeros((freq_slice,freq_slice)), cmap=cmap_2,norm=mcolors.LogNorm(vmin=1e-4, vmax=1),rasterized=True)
line13, = ax13.plot(Z[0,0,::idx_z_sample], np.zeros_like(Z[0,0,::idx_z_sample]), color='black')
vlines03 = ax03.vlines([0, 0, 0, 0], -1, 1, color=[cmap_0(255), cmap_1(255), cm.inferno(200), cmap_2(255)], linestyle='--', lw=0.8) # dummy positions
vlines13 = ax13.vlines([0, 0, 0, 0], -1, 1, color=[cmap_0(255), cmap_1(255), cm.inferno(200), cmap_2(255)], linestyle='--', lw=0.8) # dummy positions

add_colorbar(ax00, im00, r"$I_\lambda$")
add_colorbar(ax01, im01, r"$I_\lambda$")
add_colorbar(ax02, im02, r"$I_\lambda$")
add_colorbar(ax10, im10, r"$|\mathcal{V}|^2$")
add_colorbar(ax11, im11, r"$|\mathcal{V}|^2$")
add_colorbar(ax12, im12, r"$|\mathcal{V}|^2$")
cax03 = fig.add_axes([0.889, ax03.get_position().y0, 0.005, ax03.get_position().height], rasterized=True)  # Add a colorbar for ax03 **outside the grid to prevent compression**
cbar03 = fig.colorbar(im03, cax=cax03)

# Add guiding lines
line_1 = ConnectionPatch((-1, theta_ph), (2, theta_ph), 'data', 'data', axesA=ax00, axesB=ax03, color=(0.9,0.9,0.9,0.6), linewidth=1, linestyle='dotted')
line_2 = ConnectionPatch((-1, -theta_ph), (2, -theta_ph), 'data', 'data', axesA=ax00, axesB=ax03, color=(0.9,0.9,0.9,0.6), linewidth=1, linestyle='dotted')
line_3 = ConnectionPatch((0,-32+freq_slice//2), (freq_slice,-32+freq_slice//2), 'data', 'data', axesA=ax10, axesB=ax12, color=(0.9,0.9,0.9,0.6), linewidth=1, linestyle='dotted')
line_4 = ConnectionPatch((0,+32+freq_slice//2), (freq_slice,+32+freq_slice//2), 'data', 'data', axesA=ax10, axesB=ax12, color=(0.9,0.9,0.9,0.6), linewidth=1, linestyle='dotted')
fig.add_artist(line_1); fig.add_artist(line_2); fig.add_artist(line_3); fig.add_artist(line_4);

# Add observer's eye
ax03.add_artist(eye_triangle);
ax03.add_patch(eye_ball);
ax03.add_patch(eye_pupil);
ax03.add_patch(pupil);

# Function to update each frame
def update(frame):
    t = t_values[frame]
    
    # Interpolate for time t
    grf_int = grf_interpolate(grf, physical_size, t, X, Y, Z)
    
    # Compute images and visibilities
    idx_z_0 = idx_mid_proj - int(1 * n_idx * t)
    idx_z_1 = idx_mid_proj - int(0.20 * n_idx * t)
    idx_z_2 = idx_mid_proj + int(0.1 * n_idx * t)

    # Compute images and visibilities
    I_lambda_total_0 = np.abs(I_lambda_total(X[:,:,idx_z_0], Y[:,:,idx_z_0], Z[:,:,idx_z_0], grf_int[:,:,idx_z_0], tau_0, delta_tau, spectral_index, theta_ph,t))
    I_lambda_total_1 = np.abs(I_lambda_total(X[:,:,idx_z_1], Y[:,:,idx_z_1], Z[:,:,idx_z_1], grf_int[:,:,idx_z_1], tau_0, delta_tau, spectral_index, theta_ph,t))
    I_lambda_total_2 = np.abs(I_lambda_total(X[:,:,idx_z_2], Y[:,:,idx_z_2], Z[:,:,idx_z_2], grf_int[:,:,idx_z_2], tau_0, delta_tau, spectral_index, theta_ph,t))
    I_lambda_ph_side = I_lambda_ph(np.sqrt(Z[idx_mid_side,:,:]**2 + Y[idx_mid_side,:,:]**2), theta_ph)
    I_lambda_emission_side = I_lambda_emission(X[idx_mid_side,:,:], Y[idx_mid_side,:,:], Z[idx_mid_side,:,:], grf_int[idx_mid_side,:,:],tau_0, delta_tau, spectral_index, theta_ph,t, False)
    I_lambda_total_3 = I_lambda_ph_side + I_lambda_emission_side

    I_lambda_spectrum = I_lambda_total(X[:,:,::idx_z_sample], Y[:,:,::idx_z_sample], Z[:,:,::idx_z_sample], grf_int[:,:,::idx_z_sample], tau_0, delta_tau, spectral_index, theta_ph,t)
    f_lambda = np.sum(I_lambda_spectrum, axis=(0,1))
    f_lambda /= f_lambda[0]

    I_lambda_total_0_padded = np.pad(I_lambda_total_0, n_pad, mode='constant', constant_values=0) # pad the image for better fourier resolution
    I_lambda_total_1_padded = np.pad(I_lambda_total_1, n_pad, mode='constant', constant_values=0) # pad the image for better fourier resolution
    I_lambda_total_2_padded = np.pad(I_lambda_total_2, n_pad, mode='constant', constant_values=0) # pad the image for better fourier resolution
    visibility_0 = np.abs(fftshift(fftn(I_lambda_total_0_padded))[n_mid-(freq_slice//2):n_mid+(freq_slice//2)+1,n_mid-(freq_slice//2):n_mid+(freq_slice//2)+1])**2
    visibility_0 = visibility_0 / np.max(visibility_0)
    visibility_1 = np.abs(fftshift(fftn(I_lambda_total_1_padded))[n_mid-(freq_slice//2):n_mid+(freq_slice//2)+1,n_mid-(freq_slice//2):n_mid+(freq_slice//2)+1])**2
    visibility_1 = visibility_1 / np.max(visibility_1)
    visibility_2 = np.abs(fftshift(fftn(I_lambda_total_2_padded))[n_mid-(freq_slice//2):n_mid+(freq_slice//2)+1,n_mid-(freq_slice//2):n_mid+(freq_slice//2)+1])**2
    visibility_2 = visibility_2 / np.max(visibility_2)

    # Update plots
    im00.set_array(I_lambda_total_0.ravel())
    im01.set_array(I_lambda_total_1.ravel())
    im02.set_array(I_lambda_total_2.ravel())
    im03.set_array(I_lambda_total_3)
    im10.set_array(visibility_0)
    im11.set_array(visibility_1)
    im12.set_array(visibility_2)
    line13.set_ydata(f_lambda)
    ax13.set_title(f"$t = {t:.2f}$")

    # Update ticks and lines
    z_ticks = [z_lin[idx_z_0], z_lin[idx_z_1], 0, z_lin[idx_z_2]]
    new_segments = [[[z, -1], [z, 2]] for z in z_ticks]
    vlines03.set_segments(new_segments)
    vlines13.set_segments(new_segments)
    ax13.set_xticks(z_ticks)
    ax13.set_xticklabels([r'$\lambda_1$', r'$\lambda_2$', r'$\lambda_\mathrm{rest}$', r'$\lambda_3$'])

    return [im00, im01, im02, im10, im11, im12, line13]

# Create animation
ani = animation.FuncAnimation(fig, update, frames=len(t_values), interval=100, blit=False)

# Save animation
if bool_save:
    ani.save("EEM_animation.mp4", writer="ffmpeg", dpi=300)
    fig.savefig("EEM_animation.pdf", dpi=300, bbox_inches='tight')
    fig.savefig("EEM_animation.png", dpi=300, bbox_inches='tight')

plt.show()
```
