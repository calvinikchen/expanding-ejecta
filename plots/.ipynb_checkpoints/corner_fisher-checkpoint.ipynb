```python
cov_spec = np.genfromtxt('cov_spec_plot/cov_spec_a2_t2.csv', delimiter = ',')
print(np.all(np.linalg.eig(cov_spec)[0] > 0))

tmp = np.random.multivariate_normal(initial_spec ,cov_spec, N)
data_spec = np.zeros((N,7))

data_spec[:,0] = tmp[:,0]
data_spec[:,2:6] = tmp[:,1:5]

#data_spec[:,1] = np.random.random(N)
#data_spec[:,6] = np.random.random(N)
```

```python
inv_cov_sp = np.zeros((9,9))
inv_cov_sp[:7,:7] = np.linalg.inv(cov_spec)
```

```python
cov_3 = np.linalg.inv(inv_cov_ck + inv_cov_sp)# - np.eye(9)*1e-1)
mu_3  = jnp.array([0, 0, 0, np.log(1.2), np.pi/8*4/2, 0,0,0,0])

data_3 = np.random.multivariate_normal(mu_3, cov_3, size = N)

data_9 = np.zeros(data_3.shape)
data_9[:,0] = data_3[:,0]
data_9[:,1] = data_3[:,7]
data_9[:,2:6] = data_3[:,1:5]
data_9[:,6] = data_3[:,8]
#data_9[:,5] = np.log(data_9[:,5])
data_9 = data_9[:,:7]
```

```python
ds_ar = np.logspace(0, 1, 11)
ind = 7
ind_e = 3

#print(ds_ar)
reg = np.zeros(11)
reg_eta = np.zeros(11)
for i in range(11):
    fac = ds_ar[i]**4
    cov_3 = np.linalg.inv(inv_cov_ck/fac + inv_cov_sp/fac)# - np.eye(9)*1e-1)
    reg[i] = np.sqrt(np.diag(cov_3)[ind])#*10**(0.4*(12-9.5))
    reg_eta[i] = np.sqrt(np.diag(cov_3)[ind_e])

reg_sqrt = np.zeros(11)
reg_sqrt_eta = np.zeros(11)
for i in range(11):
    fac = ds_ar[i]**2
    cov_3 = np.linalg.inv(inv_cov_ck/fac**2 + inv_cov_sp/fac)# - np.eye(9)*1e-1)
    reg_sqrt[i] = np.sqrt(np.diag(cov_3)[ind])#*10**(0.4*(12-9.5))
    reg_sqrt_eta[i] = np.sqrt(np.diag(cov_3)[ind_e])

reg_fix = np.zeros(11)
reg_fix_eta = np.zeros(11)
for i in range(11):
    fac = ds_ar[i]**4
    cov_3 = np.linalg.inv(inv_cov_ck/fac + inv_cov_sp)# - np.eye(9)*1e-1)
    reg_fix[i] = np.sqrt(np.diag(cov_3)[ind])#*10**(0.4*(12-9.5))
    reg_fix_eta[i] = np.sqrt(np.diag(cov_3)[ind_e])


prior = np.diag([0, 0, 0, 1/0.2**2, 0, 0, 0, 0, 0])

reg_prior = np.zeros(11)
reg_prior_eta = np.zeros(11)
for i in range(11):
    fac = ds_ar[i]**4
    cov_3 = np.linalg.inv(inv_cov_ck/fac + inv_cov_sp/fac + prior)# - np.eye(9)*1e-1)
    reg_prior[i] = np.sqrt(np.diag(cov_3)[ind])#*10**(0.4*(12-9.5))
    reg_prior_eta[i] = np.sqrt(np.diag(cov_3)[ind_e])
```

```python
plt.figure(figsize = (12, 5))
plt.subplot(121)
plt.plot(ds_ar, reg)
plt.plot(ds_ar, reg_sqrt)
plt.plot(ds_ar, reg_fix)
plt.plot(ds_ar, reg_prior)
plt.xscale('log')
plt.yscale('log')
plt.xlabel(r'$D$')
plt.ylabel(r'$\sigma_D/D$')

plt.subplot(122)
plt.plot(ds_ar, reg_eta, label = r'$\sigma_\text{sp}\propto F$')
plt.plot(ds_ar, reg_sqrt_eta, label = r'$\sigma_\text{sp}\propto \sqrt{F}$')
plt.plot(ds_ar, reg_fix_eta, label = r'$\sigma_\text{sp} = \text{const.}$')
plt.plot(ds_ar, reg_prior_eta, label = r'$\sigma_\text{sp}\propto F + \text{prior}$')
plt.xscale('log')
plt.yscale('log')
plt.xlabel(r'$D$')
plt.ylabel(r'$\sigma_\eta/\eta$')
plt.legend()
plt.title(r'$\eta = 1$')
#plt.savefig('spectrum_sigma.png', bbox_inches = 'tight')
```
